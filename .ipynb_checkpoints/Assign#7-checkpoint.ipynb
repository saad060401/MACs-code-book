```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt

# download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# read review text file with encoding
with open("Ring Camera 100 reviews.txt", "r", encoding="utf-8") as file:
    reviews = file.read()

# tokenize text into words
words = word_tokenize(reviews)

# remove stop words
stop_words = set(stopwords.words('english')) #allows for fast lookup ops after conversion to data structure
# below is a check for word in words to be lower case and being in the alphabet
# if the 2 conditions are met they will be part of filtered words list
filtered_words = [word.lower() for word in words if word.lower() not in stop_words and word.isalpha()]

# count frequency of each word
word_freq = Counter(filtered_words)

# plot top 20 most common words
# returns a list of the most common words into top_words
top_words = word_freq.most_common(20)
# unpacks the list by using the function zip
# x contains the words and y contains the freq 
x, y = zip(*top_words)
plt.figure(figsize=(10, 6))
plt.bar(x, y)
plt.title('20 Top Most Common Words in Ring Camera Amazon Reviews')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# specify keyword
keyword = "quality"

# display and interpret sentences containing the keyword
for sentence in sentences:
    if keyword in sentence.lower():
        print(f"Sentence: {sentence}\n")

```
